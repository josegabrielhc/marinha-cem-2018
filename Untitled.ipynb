# Extração de dados do concurso de Corpo de Engenheiros da Marinha
Os dados foram copiados da tabela do pdf diretamente para arquivos de texto nomeados de acordo com o curso escolhido para a prova (do concurso).

```python
import os
import pandas as pd
```

Listar os arquivos txts gerados manualmente com os resultados do concurso

```python
files = [i for i in os.listdir('.') if i.endswith('.txt')]
```

Cria a partir dos arquivos txt arquivos formatados csv. Poderia gerar o dataframe, mas preferi exportar como csv

```python
for j in range(len(files)):
    file = open(files[j], encoding='utf8')
    lines = file.readlines()
    lines = [i.replace('\n','') if not i.endswith(' \n') else i.replace('\n','')[:-2] for i in lines]
    values = [i.split(' ') for i in lines]
    final = []
    saida = open(files[j].replace('.txt','.csv'),'w',encoding='utf8')
    saida.write('inscricao;nome;nota0;nota1;data_nasc;org;curso\n')
    for value in values:
        insc = value[0]
        org = value[-1]
        data_nasc = value[-2]
        nota1 = value[-3]
        nota0 = value[-4]
        nome = ' '.join(value[1:-4])
        #print(insc, nome, nota0, nota1, data_nasc, org)
        saida.write('%s;%s;%s;%s;%s;%s;%s\n'%(insc, nome, nota0.replace(',','.'), 
                                              nota1.replace(',','.'), data_nasc, org, files[j].replace('.txt','')))
    saida.close()
```

Update da lista de arquivos. Poderia fazer equivalente com a biblioteca OS, acima. Creio que renomear seja mais rápido.

```python
files = [i.replace('.txt','.csv') for i in files]
```

Cria um único dataframe

```python
lst = []
for i in files:
    df = pd.read_csv(i,sep=';')
    lst.append(df)
frame = pd.concat(lst, axis = 0, ignore_index = True)
```

Aprovados, a princípio, para EVC em Brasília

```python
frame.query("nota0>10 and nota1>40 and org=='Com7ºDN'")
```

Aprovados em um curso específico:

```python
frame.query("nota0>10 and nota1>=40 and curso == 'mecatronica'")
```
